```python
from PIL import ImageFont, ImageDraw, Image
from matplotlib import pyplot as plt
from numpy import array

import cv2 as cv
import re # regular expression
import os
import pytesseract

pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'  # The binary path into your code
textSearched = 'processamento'



listFiles = []

myDir = os.getcwd()
dirFolder = os.path.join(myDir, 'ARTICLES')

for file in os.listdir(dirFolder):
    filePath = os.path.join(dirFolder, file)
    if os.path.isfile(filePath): 
        listFiles.append(filePath)
        print(filePath)
```

```python
fonte = 'FONTE\calibri.ttf'


def writeText(texto, x, y, img, fonte, tamanho_texto=20):
  fonte = ImageFont.truetype(fonte, tamanho_texto)
  img_pil = Image.fromarray(img)
  draw = ImageDraw.Draw(img_pil)
  draw.text((x-10, y - tamanho_texto), texto, font = fonte, fill=(255,0,0,255))
  img = array(img_pil)
  return img


def renderBox(image, x, y, width, height, color, text):
    cv.rectangle(image, (x, y), (x + width, y + height), color)
    #cv.putText(image, text, (x-10, y-10), cv.FONT_HERSHEY_SIMPLEX, .5, (0,0,255))
    return writeText(text, x, y, image, fonte) 
    
def showImage(img):
    fig = plt.gcf()
    fig.set_size_inches(5, 5) # Define image size
    plt.axis("off")  # Remove axis X, Y to plt
    imageRgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(imageRgb)
    plt.show()

def runOcr(listFiles, textsch: str = None, color = (0, 0, 255), showTextConsole = True, saveImage=False):
    dirName, textToSave, textToShow = '', '', ''
    minConf = 60
    countOcour = 0
    
    for file in listFiles:
        image = cv.imread(file)

        result = pytesseract.image_to_data(image, lang='por', output_type=pytesseract.Output.DICT)

        titulo = str(os.path.split(file)[-1]).upper()
        textToSave +=   '\n ***********' + titulo + '***********\n' # Image name as title


        for i in range(len(result['text'])):
            if int(result['conf'][i]) > minConf:

                x = result['left']  [i] 
                y = result['top']   [i]
                width  = result['width'] [i]
                height = result['height'][i]
                confianca = str(result['conf'][i]) + '%' + ' >> ' 
                text = str(result['text'][i])

                if textsch is None:  # Find all texts

                    image = renderBox(image, x, y, width, height, color, text)    
                else:
                    if textsch.upper() in text.upper():  # Find only *** textSearched *** text.
                        countOcour +=1
                        image = renderBox(image, x, y, width, height, color, text)

                     
                textToShow += text #+ confidence + '\n' 
                textToSave += textToShow 

        if showTextConsole is True:            
            print(textToShow)
            textToShow = ''
            textToSave += "\n ------------------------------------------------------------------------------------\n\n"

        showImage(image)
        if countOcour > 0:  # If has some text
            print(f'{countOcour} ocorrência encontradas da palava {textsch}.')

            if saveImage is True:
                cv.imwrite(os.path.join(dirName, os.path.split(file)[-1]), image)
            countOcour = 0
                        
        if saveImage is True:
            dirName = (os.path.join(os.getcwd(), 'photosReaded'))

            if not os.path.exists(dirName):
                os.makedirs(dirName, exist_ok=True)
                     
    return textToSave

textToSave = runOcr(listFiles, textSearched, showTextConsole=True, saveImage=True)   
```

```python
nameFile = os.getcwd() + '\Extracao_text.txt'

f = open(nameFile, 'a')
f.write(textToSave)
f.close()
```
